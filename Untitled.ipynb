```python
import cv2
import mediapipe as mp
import math
import numpy as np
from pycaw.pycaw import AudioUtilities, IAudioEndpointVolume
from ctypes import cast, POINTER
from comtypes import CLSCTX_ALL
device=AudioUtilities.GetSpeakers()
interface=device.Activate(
IAudioEndpointVolume._iid_,CLSCTX_ALL, None)
vol= cast(interface,POINTER(IAudioEndpointVolume))
#print(vol.GetMute())
#print(vol.GetMasterVolumeLevel())
volumeR=vol.GetVolumeRange()
minvol=volumeR[0]
maxvol=volumeR[1]
vol.SetMasterVolumeLevel(0.0,None)
mp_draw=mp.solutions.drawing_utils
mp_hands=mp.solutions.hands
hand = mp_hands.Hands()
img=cv2.VideoCapture(0)
while True:
    succ,op= img.read()
    
    
    result=hand.process(cv2.cvtColor(op,cv2.COLOR_BGR2RGB))
    if result.multi_hand_landmarks:
        for i in result.multi_hand_landmarks:
            list1=[]
            for id,lm in enumerate(i.landmark):
                h,w,c=op.shape
            
                cx,cy=int(lm.x*w),int(lm.y*h)
                list1.append([id,cx,cy])
                #print(list1)
                # print(id,lm)
                mp_draw.draw_landmarks(op,i,mp_hands.HAND_CONNECTIONS)
            if list1:
                x1,y1=list1[4][1],list1[4][2]
                x2,y2=list1[8][1],list1[8][2]
                cv2.circle(op,(x1,y1),10,(155,121,12),cv2.FILLED)
                cv2.circle(op,(x2,y2),10,(155,121,12),cv2.FILLED)
                cv2.line(op,(x1,y1),(x2,y2),(1,12,12),3)
                len=math.hypot(x2-x1,y2-y1)
                volume=np.interp(len,[50,300],[minvol,maxvol])
                volBar=np.interp(len,[50,300],[400,150])
                volPer=np.interp(len,[50,300,[0,100]])
                vol.SetMasterVolumeLevel(volume,None)
                cv2.rectangle(op,(50,150),(85,400),(211,12,12),3)
                cv2.rectangle(op,(50,int9volBar),(85,400),(211,12,12),cv2.FILLED)
                cv2.putText(op,str(int(volPer)),(40,450),cv2.FONT_HERSHEY_PLAIN,4 , (24,34,34) , 3)
               
                #print(len)
    
        cv2.imshow("Image Captured",op)
        cv2.waitKey(1)
    
```
